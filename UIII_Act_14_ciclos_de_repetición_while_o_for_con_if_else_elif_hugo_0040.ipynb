#### problema:Una empresa les paga a sus empleados con base en las horas trabajadas en la semana. Para esto, se registran los días que trabajó y las horas de cada día. Realice un algoritmo para determinar el sueldo semanal de N trabajadores y además calcule cuánto pagó la empresa por los N empleados.

```python
total_empresa = 0

# Validación simple con while
n = int(input("¿Cuántos trabajadores hay?: "))

while n <= 0:
    print("❌ Número inválido")
    n = int(input("Ingresa un número válido de trabajadores: "))

i = 1

# Ciclo while para trabajadores
while i <= n:
    print("\nTrabajador", i)

    dias = int(input("¿Cuántos días trabajó?: "))
    sueldo = 0

    d = 1

    # Ciclo while para días
    while d <= dias:
        horas = float(input("Horas trabajadas en el día " + str(d) + ": "))
        pago_hora = float(input("Pago por hora: "))

        # Condiciones con if, elif, else
        if horas <= 8:
            pago_dia = horas * pago_hora
        elif horas > 8 and horas <= 12:
            extra = horas - 8
            pago_dia = (8 * pago_hora) + (extra * pago_hora * 1.5)
        else:
            extra = horas - 8
            pago_dia = (8 * pago_hora) + (extra * pago_hora * 2)

        sueldo = sueldo + pago_dia

        d = d + 1

    print("💰 Sueldo semanal del trabajador", i, "es:", sueldo)

    total_empresa = total_empresa + sueldo

    i = i + 1

# Resultado final
print("\n💵 Total pagado por la empresa:", total_empresa)
```
